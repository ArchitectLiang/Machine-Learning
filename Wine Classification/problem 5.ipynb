```python
import numpy as np
import scipy.io
import scipy
import scipy.special
import matplotlib.pyplot as plt
import scipy.stats
```

```python
x, y = np.mgrid[-5:5:.1, -5:5:.1]
pos = np.dstack((x, y))
w = (abs(x)**(1/2) + abs(y)**(1/2))**2
plt.contourf(x, y, w)
plt.colorbar()
print('1a')
```

```python
x, y = np.mgrid[-5:5:.1, -5:5:.1]
pos = np.dstack((x, y))
w = (abs(x)**(1) + abs(y)**(1))**1
plt.contourf(x, y, w)
plt.colorbar()
print('1b')
```
