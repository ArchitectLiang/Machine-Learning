```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
```

```python
x, y = np.mgrid[-2:4:.01, -3:5:.01]
pos = np.dstack((x, y))
rv = scipy.stats.multivariate_normal([1, 1], [[1, 0], [0, 2]])
plt.contourf(x, y, rv.pdf(pos))
plt.colorbar()
print('3.1')
```

```python
x, y = np.mgrid[-5:3:.01, -3:7:.01]
pos = np.dstack((x, y))
rv = scipy.stats.multivariate_normal([-1, 2], [[2, 1], [1, 4]])
plt.contourf(x, y, rv.pdf(pos))
plt.colorbar()
print('3.2')
```

```python
x, y = np.mgrid[-5:5:.01, -3:5:.01]
pos = np.dstack((x, y))
rv1 = scipy.stats.multivariate_normal([0, 2], [[2, 1], [1, 1]])
rv2 = scipy.stats.multivariate_normal([2, 0], [[2, 1], [1, 1]])
plt.contourf(x, y, rv1.pdf(pos) - rv2.pdf(pos))
plt.colorbar()
print('3.3')
```

```python
x, y = np.mgrid[-5:5:.01, -3:5:.01]
pos = np.dstack((x, y))
rv1 = scipy.stats.multivariate_normal([0, 2], [[2, 1], [1, 1]])
rv2 = scipy.stats.multivariate_normal([2, 0], [[2, 1], [1, 4]])
plt.contourf(x, y, rv1.pdf(pos) - rv2.pdf(pos))
plt.colorbar()
print('3.4')
```

```python
x, y = np.mgrid[-5:5:.01, -5:5:.01]
pos = np.dstack((x, y))
rv1 = scipy.stats.multivariate_normal([1, 1], [[2, 0], [0, 1]])
rv2 = scipy.stats.multivariate_normal([-1, -1], [[2, 1], [1, 2]])
plt.contourf(x, y, rv1.pdf(pos) - rv2.pdf(pos))
plt.colorbar()
print('3.5')
```
